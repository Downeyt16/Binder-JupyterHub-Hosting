### Set Identities and DeMorgans Laws
Intersection distributes over the symmetric difference: <br>

$$A\cap \left ( B\Delta C \right )=\left ( A\cap B \right )\Delta \left ( A\cap C \right )$$ <br>

Also, $$A\Delta B=A^{c}\Delta B^{^{c}}$$ <br>

$$\emptyset^c = U \text{ and }  U^c = \emptyset $$ <br> <br>

Identity Laws: <br>

$$A\cap U=A$$ <br>

$$A\cup U=U$$ <br> <br>


Universal Bound: <br>

$$A\cap \emptyset =\emptyset$$ <br>

$$A\cup \emptyset =A$$ <br> <br>

Idempotent Laws: <br>

$$A\cap A=A$$ <br>

$$A\cup A= A$$ <br> <br>


Commutative Laws:<br>

$$A\cap B= B\cap A$$ <br>

$$A \cup B= B\cup A$$ <br> <br>


Associative Laws: <br>

$$\left ( A\cap B \right )\cap C=A\cap \left ( B\cap C \right )$$ <br>

$$\left ( A\cup B \right )\cup C=A\cup \left ( B\cup C \right )$$ <br> <br>

Distributive Laws: <br>

$$A\cap \left ( B\cup C \right )=\left ( A\cap B \right )\cup \left ( A\cap C \right )$$ <br>

$$A \cup (B \cap C) = (A \cup B) \cap (A \cup C)$$ <br> <br>


DeMorgan's Laws: <br>

$$\left ( A\cap B \right )^{c}=A^{c}\cup B^{c}$$ <br> 

$$\left ( A\cup B \right )^{c}=A^{c}\cap B^{c}$$ <br>
