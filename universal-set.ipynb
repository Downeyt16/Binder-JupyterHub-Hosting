### Universal Set

Universal set contains any relevant elements and depends on application. <br> <br>

Example: <br>

The $U$ for a days in a week is: <br>

$$U = \{\text{Sunday,Monday,Tuesday,Wednesday,Thusday,Friday,Saturday}\}$$ <br> <br>

On the other hand, $U$ for working days of the week is: <br>

$$U = \{\text{Monday,Tuesday,wednesday,Thusday,Friday}\}$$ <br> 

Note: The above universal sets may vary from one country to another. In some contries, Saturday is a working day. <br> <br> <br>
Several commonly used sets of numbers are denoted as follows: <br> <br>
$$\begin{aligned} &\mathbb{R} \text{ is the set of real numbers} \\ \\
&\mathbb{Z} \text{ is the set of integers}, \{\ldots,-2,-1,0,1,2,\ldots\} \\ \\
&\mathbb{N} \text{ is the set of natural numbers,} \{0,1,2,3,\ldots\} \\ \\
&\mathbb{P} \text{ is the set of positive integers,} \{12,3,\ldots\} \\ \\
&\mathbb{Q} \text{ is the set of rational numbers,} \left\{\frac{2}{5},\frac{5}{2},\frac{3}{7},\ldots \right\} \\ \\
&\mathbb{I} \text{ is the set of irrational numbers,} \left\{\sqrt{2}, \pi, e, \ldots\right\} \end{aligned}$$ <br> <br>

The following subsets are from the common sets listed above: <br> <br>
$$\begin{aligned}&\mathbb{Q} \subset \mathbb{R} \\ \\ 
&\mathbb{I} \subset \mathbb{R} \\ \\
&\mathbb{N} \subset \mathbb{Z} \\ \\
&\mathbb{P} \subset \mathbb{N} \\ \\
&\mathbb{P} \subset \mathbb{Z}\end{aligned}$$ <br> <br>

An empty set is a subset of every set including itself <br>
$$\emptyset \in \left \{ \emptyset \right \} \text{ and }  \emptyset \subseteq \left \{ \emptyset \right \}$$ <br> <br>

Set union is not limited to two sets. It can be extented to an infinite number of sets. <br> 

Example: <br>

Let $B = \{0,1\}, O = \{0,1,2,\ldots,7\}, \text{ and } D = \{0,1,2,\ldots,9\}$ <br>
    
Then, <br>
    
$B \cup O \cup D\ = \{0,1,2,\ldots,9\}$ <br> <br>

Set intersection is not limited to two sets, it can be extented to an infinite number of sets. <br> 

Example: <br>

Let $B = \{0,1\}, O = \{1,2,3,\ldots,7\} \text{ and } D = \{0,1,2,\ldots,9\}$ then $B \cap O\cap D = \{0,1\}$ <br> <br>

Example: <br>

Let $S_n \text{ where }n=1,2,3,\ldots$ be an infinite collection of sets. <br>

That is, $S_1,S_2,S_3,\ldots,$ are sets. <br> <br>

Their intersection is expressed as: <br> 

$\bigcup_{n}S_{n} \text{ where } n=1,2,3,\ldots$ <br> <br>

An element is in the intersection if and only if it is a member of every set. <br>

That is $x\in \bigcup_{n} S_{n}$, if and only if $x \in S_{n}$, for all $n$. <br>
