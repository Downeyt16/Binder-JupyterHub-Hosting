## Power Set

The power set is a set of all subsets of a set. It includes the null set $\varnothing$ and also the set itself. <br> 

The power set of a set $A$ is denoted by $\mathcal{P}$(A). <br> <br>

Example: <br>

Let $B = \{0,1\}$, the set of binary digits. <br>

Its power set $P(B)$ is all possible subsets including the null set $\emptyset$ and $B$ itself. <br>

$\mathcal{P}(B) = \{\emptyset,\{0\},\{1\},\{0,1\}\}$ <br>

Note that the elements of the power set are themselves sets. Therefore, the power set is a set of sets.<br> <br>

Example: <br>

Let $A = \{a,b,c\}$ be a set. Its power set is all possible subsets including the null set $\emptyset$ and $A$ itself. <br>

$\mathcal{P}(A) = \{\emptyset,\{a\},\{b\},\{c\},\{a,b\},\{a,c\},\{b,c\},\{a,b,c\}\}$ <br>

We systematically enumerate subsets of cardinality $0,1,2 \text { and } 3$. Recall that cardinality of a set is the number of elements in the set. <br>

Theorem: Let $A$ be a finite set with $|A|=n$. Then, the number of sets in its power set $\mathcal{P}(A)$ is $2^{n}$. <br>

This can be proved by induction on the number of elements of $A$. <br> <br>

<b>Base case: </b>


When $n=0$ and $|A|=0$ clearly, $A = \emptyset$ <br>

According to the induction hypothesis, $|\mathcal{P}(A)| = 2^{n} = 2^{0} = 1$ <br>

Base case is satisfied. <br> <br>


Next consider the induction step and let $|A|=n$.<br>

By the induction hypothesis,$|\mathcal{P}(A)|=2^{n}$ <br>

That is, the number of subsets of $A= 2^{n}$ <br>

Let $B = A \cup \{a\}$. That is, create a set $B$ which has all the elements of $A$ and one additional element $a$. Therefore, $|B| = n+1$  <br> <br>

Now consider all the subsets of $B$. There are two kinds : <br>

Those that are in $A$ (which do not include the new element $a$), and those that are in $A$ but include the new element $a$. <br>

In other words, numbers of subsets of $B$ equals those that are in $A$ plus those that are in $A$ plus the new element $a$.<br>

$= 2^{n}$ (by induction hypothesis) + $2^{n}$ (there are same subsets of $A$, but each susbsets now has a new element attached.) <br>

$\mathcal{P}(B) = 2^{n} + 2^{n} = 2.2^{n} = 2^{n+1}$ <br>

Hence proved <br> <br>
