## Set Partitions

Grouping elements of a set into non-empty subsets in a way that every element is included in exactly one one subset is called partitioning of a set. <br>

The intersection of any two subsets in the partition is an empty set. <br> 

In other words, the subsets in the partition are said to be pairwise disjoint.<br> <br> <br>



Example: <br>

The set $A = \{1,2,3\}$ has the following five partitions: <br>

$\{\{1\} , \{2\} , \{3\}\}$ is a partition, which is also written as $1|2|3$ <br> <br>

The other four partitions are: <br>

$\{\{1,2\},\{3\}\}$ or $1 2 | 3$ <br>

$\{\{1,3\},\{2\}\}$ or $1 3 | 2$ <br>

$\{\{1\},\{2,3\}\}$ or $1 | 3  2$ <br>

$\{\{1,2,3\}\}$ or $1 2 3$  <br>

Note : Partitions should not contain the null set, an element should appear in exactly one partition, and every element should appear in same partition. <br> <br> <br>



Formally, let $A$ be a set and $A_{1},A_{2},A_{3}, ... ,A_{n}$ be its subsets. the subsets form a partition on $A$, if the following three conditions hold: <br>

(1)  $A_{1} \cup A_{2} \cup A_{3} \cup ... \cup A_{n} = A$ <br>

(2)  $A_{i} \neq \emptyset$ for all i <br>

(3) $A_{i} \cap A_{j} = \emptyset$ for all $i \neq j$ <br> <br> <br>




Example: <br> 

Consider the following set of phone numbers (written in a format used in the United States): <br>

Phones = {252-328-9680, 704-978-4243, 336-521-1121, 704-888-0102, 336-678-1415, 252-968-1212} <br> 

We deliver an equivalence relation to decoupe the set into its equivalence classes. <br>

Therefore, the equivalence classes are: <br>

$\{252-328-9680, 252-968-1212\}$ <br>

$\{704-978-4342, 704-888-0102\}$ <br>

$\{336-521-1121, 336-678-1415\}$ <br> 

Note that all elements (i.e., phone numbers) in an equivalence class have the "same area code." <br> <br> <br>



Equivalence classes form a partition on the set Phones. <br>

The subsets of the partion are: <br>

$\{\{252-328-9680, 252-968-1212\}$, <br>

$\{704-978-4342, 704-888-0102\}$, <br>

$\{336-521-1121, 336-678-1415\} \}$ <br> <br> <br>




Confirm that above subsets are a partition by verifying: <br> <br>

$(i)$ Each phone number appears in only one subset. <br> <br>
$(ii)$ Every phone number appears in same subset. <br> <br> 
$(iii)$ No subset is a null set. <br> <br> 

We also verify that the relation "same area code" is an equivalence on the set Phones. <br> <br> <br>




A relation R on a set A is an equivalence relation if it has the following three properties: <br> <br>

$(i)$ Reflexive, if $x \in A$, then $(x,x) \in R$ <br> <br>

$(ii)$ Symmetric, if $(x,y) \in R$, then $(y,x) \in R $ <br> <br>

$(iii)$ Transitive, if $(x,y) \in R$ and if $(y,z) \in R$, then $(x,z) \in R$ <br> <br> <br>



It is easy to verify that the relation "Same area code" satisfies all three properties, hence, it is an equivalence relation. <br>

The relation is reflexive. Let $x$ be any element of the set phones. It is obvious that any phone number and itself will have the same area code. <br>

The relation is symmetric. Let $x$ and $y$ be any two elements of the set Phones, if $x$ and $y$ have the same area code, it is also true that $y$ and $x$ have the same area code. <br>

The relation is transitive. Let $x$, $y$ and $z$ be any three elements of the set phones.<br>

If $x$ and $y$ have the same area code and $y$ and $z$ have the same area code, then $x$ and $z$ must have the same area code.(by transitive properties). <br> <br> <br>



The total number of partitions of an n-element set is the Bell number, $B_{n}$ <br>

The first few Bell numbers are $B_{0} = 1,B_{1} = 1,B_{2} = 2,B_{3} = 5,B_{4} = 15$ and $B_{5} = 52$ <br> 

Bell numbers satisfies the recursion <br>

$$B_{n+1} = \sum_{k=0}^{n}\binom{n}{k}B_{k}$$ <br> 

Note: $\binom{n}{k}$ is the binomial coefficience and is given by $\binom{n}{k} = \frac{n!}{k!(n-k)!}$ <br> <br> <br>




The Stirling number of the second kind (or Stirling partition number) gives the number of ways to partition a set of $n$ objects into $k$ non-empty subsets, denoted $S(n,k)$. <br> 

$S(n,k)$ is computed as: <br>

$$s(n,k) = \frac{1}{k!}\sum_{i=0}^{k}(-1)^{i}\binom{k}{i}(k-i)^{n}$$ <br> 

A non-crossing partition of a set $A$ is a partition in which no two blocks "cross" each other. <br> <br>



If the elements $a$ and $b$ belongs to one block and elements $i$ and $j$ belongs to another block, then $a$ $i$ $b$ $j$ is a crossing partition whereas $a$ $i$ $j$ $b$ and $a$ $b$ $i$ $j$ are noncrossing. <br>

Note: $a$ $i$ $j$ $b$ is non-crossing as the arches connecting $a$ and $b$, and $i$ and $j$ do not cross each other. <br> <br>



The Catalan number, denoted $C_{n}$, gives the number of non-crossing partitions of an n-element set <br>

$$C_{n} = \frac{1}{n+1}\binom{2n}{n}$$ <br>
