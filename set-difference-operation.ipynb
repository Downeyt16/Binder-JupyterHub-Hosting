###  Set Difference Operation

The difference of two sets $A$ and $B$, denoted as $A \setminus B$, is the set of elements in $A$ but not in $B$.  The symbol, $A - B$ is also used to denote set difference. <br>

Example: <br>

Let $O = \{0, 1, 2, 3, 4, 5, 6, 7\}$ be the set of octal digits and $B = \{0, 1 \}$ be the set of binary digits. Then, <br>

$$O \setminus B = \{2, 3, 4 , 5, 6, 7\}$$ <br>

There are the elements in $O$ but not in $B$.  The Venn diagram shown below illustrates the set difference: <br>

To compute $O$ \ $B$, find elements that are common to $O$ and $B$, and remove them from $O$. <br>

More formally, <br>

$A \setminus B = \{x \mid x \in A \text{ and }  x \notin B \}$ <br> <br> <br>



### Set Symmetric Difference

The symmetric difference between two sets $A$ and $B$, denoted as $A \bigtriangleup B$ is the union of the two sets without the elements that are common to $A$ and $B$. <br>

Example:<br>

Let $A = \{1, 2, 3, 4, 5\} \text{ and } B = \{4, 5, 6, 7, 8, 9\}$. <br>

$A \bigtriangleup B = \{1, 2, 3, 6, 7, 8, 9\}$ <br>

as illustrated by  Venn Diagram below: <br>

<img src="F13.png" alt="drawing" width="333px"> <br>

$A \bigtriangleup B$ is the union of elements that are unique to each of $A$ and $B$. <br> <br>

A generic Venn diagram for the symmetric difference of $A$ and $B$ is shown below: <br>

The shaded area represents $A \bigtriangleup B$ <br>

<img src="symmetric-difference.png" alt="drawing" width="333px"> <br> 




Formally, <br>

$$\begin{aligned}A \bigtriangleup B &= \{x \mid (x \in A \text{ and }x \notin B) \text{ or } (x \notin A \text{ and } x \in B)\}\\ \\
&=(A-B) \cup (B-A)\end{aligned}$$ <br> <br>



Equivalently, we can express $\bigtriangleup$ as the union of two sets, minus their intersection. Then, <br> <br>

$$A \bigtriangleup B = (A \cup B) \setminus (A \cap B)$$ <br> <br> 


Lastly, $\bigtriangleup$ can also be expressed in terms of the **exclusive OR** operation, which is denoted as **$XOR$** or **$\oplus$.** <br> <br>

$$A \bigtriangleup B = \{ x \mid (x \in A) \oplus (x \in B) \}$$ <br>

$\oplus$ is a logical operation which outputs true only when inputs differ as shown in the truth table below: <br>
 
$x$ | $y$ |  $x \oplus y$
- | - | --
`F (False)` | `F (False)` | `F (False)`
`F (False)` | `T (True)` | `T (True)`
`T (True)` | `F (False)` | `T (True)`
`T (True)` | `T (True)` | `F (False)` 

<br> <br>The symmetric difference, $\bigtriangleup$, is communative and associative:<br> <br>

$$A \bigtriangleup B = B \bigtriangleup A$$ <br> <br>

$$(A \bigtriangleup B) \bigtriangleup C = A \bigtriangleup (B \bigtriangleup C)$$ <br> <br>

Also, <br> <br>
$$A \bigtriangleup \emptyset = A$$ <br>

$$A \bigtriangleup A = \emptyset$$ <br>
