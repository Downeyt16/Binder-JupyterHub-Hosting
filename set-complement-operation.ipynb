### Set Complement Operation

Complement of a set represents all the elements not in the set. In other words, the complement of set $A$ are those elements in the universal set $U$ but not in $A$. <br>

The complement of $A$ is denoted by $A^{c}$.<br>

In the figure below, the shaded area represents $A^{c}$. <br>
<img src="F7.png" alt="drawing" width="333px"> <br> 

More formally, <br> 

$$A^{c}=\{ x\mid x \in U \text{ and } x \notin A \}$$ <br>

Alternative notation for $A^{c}$ is $\bar{A}$ <br> <br> <br>




Example: <br>

$\mathbb{Z}$ is the set of integers. <br>

Assume $U = \mathbb{Z}$. That is, the universal set is the set of integers. <br>

Let $E = \{\text{set of even numbers}\}$. <br>

Then $E^{c} = \{\text{the set of odd numbers}\}$ , since the universal set is made of both even and odd numbers. <br><br> <br> <br>



Example: <br>

Let $U = \{\mathbb{Z}\}$ <br>

Let $M = \{x \mid x \in \mathbb{Z} \text{ and } (3\mid x)\}$ be the set of integers that are multiples of $3$. <br>

In other words, $M = \{\ldots,-9,-6,-3,0,3,6,9,\ldots\}$ <br>

What is the compliment of $M$? <br>

$M^c = \{ x \mid x \in \mathbb{Z} \text{ and } (3\nmid x)\}$ <br>

$M^c = \{\ldots,-10,-8,-7,-5,-4,-2,-1,0,1,2,4,5,7,8,10,\ldots\}$ is the set of integers that are not multiples of 3. <br> <br> <br> 



Example: <br>

What is complement of a complement? <br>

In other words, what is $(A^c)^c$, where $A$ is a set. <br>

$A^c$ is the set of all elements that are not in $A$ (but are in $U$) as shown below in the Venn diagram(shaded area)
<img src="F8.png" alt="drawing" width="333px"> <br> 

Next, the complement of the shaded area is the set of all elements not in the shaded area, which is the set $A$ <br> 

Therefore $( A^c)^c = A$ <br> <br> <br> 



Example: <br>

What is the complement of $U$? <br>

Since $U$ is the universal set, it contains all the elements. There are no elements outside it. In other words, $U^c =\{\text{ a set with no elements}\} = \phi =$ the null set. <br>
